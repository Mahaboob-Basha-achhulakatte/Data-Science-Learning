```python
print("Hello, World!")
print("=====VARIABLES=====")
name = "basha"
age = 22
print("name")
print("age")
print()

print("=====DATA TYPES=====")
name = "basha" #string
age = 22 #int
weight = 50.5 #float
student = True #boolean
print(type(name))
print(type(age))
print(type(weight))
print(type(student))
print()


print("=====OPERATORS=====")
x = 10
y = 5
print("addition:", x + y)
print("subtraction:", x - y)
print("multiplication:", x * y)
print("division:", x / y)
print("modulus:", x % y)
print("exponentiation:", x ** y)
print()


print("=====COMPARISON OPERATORS=====")
x = 10
y = 5
print("equal:", x == y)
print("not equal:", x != y)
print("greater than:", x > y)
print("less than:", x < y)
print("greater than or equal to:", x >= y)
print("less than or equal to:", x <= y)
print()


print("=====IF-ELSE STATEMENTS=====")
age = 18
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
print()

print("=====nested if else statements=====")
age = 18
if age >= 18:
    print("You are an adult.")
    if age >= 65:
        print("You are also a senior citizen.")
else:
    print("You are a minor.")
print()


print("=====LOOPS=====")
for i in range(5):
    print(i)
print()

print("=====WHILE LOOP=====")
count = 0
while count < 5:
    print(count)
    count += 1
print()

print("=====FUNCTIONS=====")
def greet(name):
    print("Hello, " + name + "!")
greet("name")

print()
print("=====LISTS=====")
fruits = ["apple", "banana", "orange"]
print(fruits)
print(fruits[0])
print()


print("=====SETS=====")
unique_numbers = {1, 2, 3, 4, 5}
print(unique_numbers)
print()

print("=====TUPLES=====")
#tuples
coordinates = (10, 20)
print(coordinates)
print()


print("=====STRINGS=====")
message = "Hello, World!"
print(message)
print(len(message))
print(message.upper())
print(message.lower())
print()



name = "basha"
age = 22
print("name:", name)
print("age:", age)
print()
a=10
b=5
print("addition:", a + b)
print()
for i in range(5):
    print(i)
print()
def hello():
    print("Hello, World!")
hello()



```
