```python
#======================
# basic pandas
#======================

import pandas as pd
print("Pandas imported success")


#====================
# load data sets
#====================
print("========load datasets=======")
df=pd.read_csv("quikr_car.csv")

#========================
# Display the first 5 rows of the dataset
#========================
print("========Display the first 5 rows of the dataset=======")
print("\n1. FIRST 5 ROWS")
print(df.head())

#==========================
# Display the last 5 rows
#===========================
print("========Display the last 5 rows of the dataset=======")
print("\n2 . LAST 5 ROWS")
print(df.tail())

#==========================
# Display column names of the dataset
#===========================
print("========Display column names of the dataset=======")
print("\n3. COLUMN NAMES")
print(df.columns)

#==========================
# Display the shape of the dataset
#===========================
print("========Display the shape of the dataset=======")
print("\n4. SHAPE")
print(df.shape)


#==========================
# Display dataset information
#===========================
print("========Display dataset information=======")
print("\n5. DATASET INFORMATION")
print(df.info())


#==========================
#Display data type
#===========================
print("========Display data type=======")
print("\n6. DATA TYPE")
print(df.dtypes)


#==========================
# Display missing values in the dataset
#===========================
print("========Display missing values in the dataset=======")
print("\n7. MISSING VALUES")
print(df.isnull().sum())


#==========================
# Display summary statistics of the dataset
#===========================
print("========Display summary statistics of the dataset=======")
print("\n8. SUMMARY STATISTICS")
print(df.describe())

print("\n============END OF THE PROGRAM============")
```
